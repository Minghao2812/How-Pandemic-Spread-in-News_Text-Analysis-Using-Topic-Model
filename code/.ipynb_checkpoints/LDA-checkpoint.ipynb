### Make corpus
Corpus is a list of lists(each document as a sub-list)

```python
def tokenize(text):
    # Remove punctuation
    tokenizer = nltk.RegexpTokenizer(r"\w+|\S")
    tokens = tokenizer.tokenize(text)

    # Remove Stop Words (all lowercase in corpora)
    stop_words = nltk.corpus.stopwords.words("english")
    stop_words.extend(range(0, 1000))
    stop_words.extend(
        [
            ",",
            ".",
            "-",
            "–",
            "‘",
            "’",
            "“",
            "”",
            "'",
            "?",
            '"',
            "!",
            "(",
            ")",
            ":",
            "@",
            "[",
            "]",
            "&",
            "*",
        ]
    )
    stop_words.extend(
        [
            "said",
            "says",
            "go",
            "went",
            "gone",
            "else",
            "taken",
            "per",
            "cent",
            "people",
            "also",
            "year",
            "would",
            "could",
            "many",
            "make",
            "get",
            "like",
            "dont",
            "ya",
            "one",
            "virus",
            "000",
        ]
    )
    filteredTokens = [
        w.lower() for w in tokens if not w.lower() in stop_words
    ]  # if w is not a stop word, then append it into filteredTokens

    return filteredTokens
```

### The dominant topic of each document (article or comment)

```python
def doc_dom_topic(lda, all_df_ord, corpus, text):
    """
    Get the dominant topic of each document (article or comment).
    All df, corpus, text are following the same order.
    """
    dom_topic_df = copy.deepcopy(all_df_ord)  # all_df_ord
    dom_topic_df["dominant_topic"] = None
    dom_topic_df["dominant_topic_probability"] = None
    for i, row in enumerate(lda[corpus]):
        # sort topics by their probability
        dom_topic = sorted(row, key=lambda x: (x[1]), reverse=True)[0][0]
        dom_topic_pro = sorted(row, key=lambda x: (x[1]), reverse=True)[0][1]
        dom_topic_df["dominant_topic"][i] = dom_topic
        dom_topic_df["dominant_topic_probability"][i] = dom_topic_pro
    return dom_topic_df


dom_topic_df = doc_dom_topic(optimal_lda, all_df_ord, corpus_all_ord, all_text_ord)
dom_topic_df
```

```python
def doc_all_topic(lda, all_df_ord, corpus, text):
    """
    Get all topics of each document (article or comment) in a probability descending order.
    All df, corpus, text are following the same order.
    """
    all_topic_df = copy.deepcopy(all_df_ord)  # all_df_ord
    all_topic_df["topics"] = None
    for i, row in enumerate(lda[corpus]):
        # sort topics by their probability
        topic_list = [tup[0] for tup in sorted(row, key=lambda x: (x[1]), reverse=True)]
        all_topic_df["topics"][i] = topic_list
    return all_topic_df


all_topic_df = doc_all_topic(optimal_lda, all_df_ord, corpus_all_ord, all_text_ord)
all_topic_df.to_csv("all_topic.csv", encoding="utf-8-sig")
all_topic_df
```

```python
def topic_sim(list1, list2):
    if len(list1) == len(list2):
        similarity = 1 - cosine(list1, list2)
    else:
        short = min(len(list1), len(list2))
        similarity = 1 - cosine(list1[:short], list2[:short])
    return similarity
```

```python
similarity_df = pd.DataFrame()
similarity_df["news_id"] = similarity_df["num_comments"] = similarity_df[
    "avg_similarity"
] = None
similarity_df
```

```python
for i, row in (all_topic_df.loc[all_topic_df["type"] == 0]).iterrows():
    comments_counter = 0
    sim_list = []
    news_id = row["news_id"]
    article_topics = row["topics"]
    comments_sub_df = all_topic_df.loc[all_topic_df["news_id"] == news_id][
        all_topic_df["type"] == 1
    ]
    if (
        len(comments_sub_df) > 0
    ):  # make sure there is at least one comment to that article
        for j, rrow in comments_sub_df.iterrows():
            comment_topics = rrow["topics"]
            sim_list.append(topic_sim(article_topics, comment_topics))
            comments_counter += 1
        similarity_df.loc[i] = [
            int(news_id),
            int(comments_counter),
            (sum(sim_list) / len(sim_list)),
        ]
```

```python
similarity_df.sort_values(by="avg_similarity", inplace=True)
similarity_df.to_csv("similarity.csv", encoding="utf-8-sig")
similarity_df
```

```python
def time_topic(lda, time_df,corpus, text):
    """
    Get all topics of each document (article or comment) in a probability descending order.
    All df, corpus, text are following the same order.
    """
    all_topic_df = copy.deepcopy(time_df)  # all_df_ord
    all_topic_df["topics"] = None
    for i, row in enumerate(lda[corpus]):
        # sort topics by their probability
        topic_list = [tup[0] for tup in sorted(row, key=lambda x: (x[1]), reverse=True)]
        all_topic_df["topics"][i] = topic_list
    return all_topic_df


all_topic_df = doc_all_topic(optimal_lda, all_df_ord, corpus_all_ord, all_text_ord)
all_topic_df.to_csv("all_topic.csv", encoding="utf-8-sig")
all_topic_df
```
